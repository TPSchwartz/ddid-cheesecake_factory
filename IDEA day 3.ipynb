```python
import pandas as pd
import matplotlib as plt
import re
import numpy as np
import datetime
```

```python
exit_notes_df = pd.read_excel(r'C:\Users\David Chaniott\Documents\Nashville Software School\python\projects\ddid-cheesecake_factory\data/idea.xlsx', sheet_name = 'Exit Notes')
```

```python
# Check for text 'Reopened'
reopened_df = exit_notes_df[exit_notes_df['Exit Note'].str.contains('reopened:', case=False, na=False)].reset_index(drop=True)
```

```python
# Drop Unused Columns
reopened_df = reopened_df.drop(columns=['POE', 'DOB', 'Notification Date', 'Initial Eligibility Date'])
```

```python
# findall. 'Reopened: ##/##/####'
reopened_df['reopened_dates'] = reopened_df['Exit Note'].str.findall('Reopened: \d{2}/\d{2}/\d{4}')
```

```python
reopened_df['reopened_dates'] = reopened_df['Exit Note'].str.findall('Reopened: \\d{2}/\\d{2}/\\d{4}')
reopened_df['exit_dates'] = reopened_df['Exit Note'].str.findall('Exit Date: \\d{2}/\\d{2}/\\d{4}')

# Creates two new Dataframes
# <reopen_date_df> & <exit_date_df>
reopen_date_df = reopened_df.reopened_dates.to_frame().astype(str)
exit_date_df = reopened_df.exit_dates.to_frame().astype(str)

#Splits the columns
reopen_date_df = reopen_date_df.reopened_dates.str.split(',', expand=True)
exit_date_df = exit_date_df.exit_dates.str.split(',', expand=True)

# Names the Columns
reopen_date_df.columns = ['reopen1', 'reopen2', 'reopen3', 'reopen4', 'reopen5', 'reopen6']
exit_date_df.columns = ['exit1', 'exit2', 'exit3', 'exit4', 'exit5', 'exit6']

# Creates <cols_r> & <cols_e> then strips text from them
cols_r = ['reopen1', 'reopen2', 'reopen3', 'reopen4', 'reopen5', 'reopen6']
cols_e = ['exit1', 'exit2', 'exit3', 'exit4', 'exit5', 'exit6']
reopen_date_df[cols_r] = reopen_date_df[cols_r].replace({'\[':'', '\]':'', '\'':'', 'Reopened:':''}, regex=True)
exit_date_df[cols_e] = exit_date_df[cols_e].replace({'\[':'', '\]':'', '\'':'', 'Exit Date:':''}, regex=True)
```

```python
# Convert Data Types to datetime
reopen_date_df = reopen_date_df.astype('datetime64')
exit_date_df = exit_date_df.astype('datetime64')
```

```python
# Join exit and reopen
reopen_exit_date_df = reopen_date_df.join(exit_date_df)
```

```python
# Assign new column with 0 'days'
reopen_exit_date_df['days1'] = np.timedelta64(0)
reopen_exit_date_df['days2'] = np.timedelta64(0)
reopen_exit_date_df['days3'] = np.timedelta64(0)
reopen_exit_date_df['days4'] = np.timedelta64(0)
reopen_exit_date_df['days5'] = np.timedelta64(0)
reopen_exit_date_df['days6'] = np.timedelta64(0)
```

```python
def awesome(R,E) :
    
    difference = (R - E)
    return(difference)
```

```python
for index, row in reopen_exit_date_df.iterrows() :
    
    #days1
    
    r = row['reopen1']
    e = row['exit1']
    reopen_exit_date_df.at[index, 'days1'] = awesome(r,e)
    
    #days2
    
    r = row['reopen2']
    e = row['exit2']
    reopen_exit_date_df.at[index, 'days2'] = awesome(r,e)
    
    #days3
    
    r = row['reopen3']
    e = row['exit3']
    reopen_exit_date_df.at[index, 'days3'] = awesome(r,e)
    
    #days4
    
    r = row['reopen4']
    e = row['exit4']
    reopen_exit_date_df.at[index, 'days4'] = awesome(r,e)
    
    #days5
    
    r = row['reopen5']
    e = row['exit5']
    reopen_exit_date_df.at[index, 'days5'] = awesome(r,e)
    
    #days6
    
    r = row['reopen6']
    e = row['exit6']
    reopen_exit_date_df.at[index, 'days6'] = awesome(r,e)
```

```python
days_only_df = reopen_exit_date_df.drop(columns=['reopen1', 'reopen2', 'reopen3', 'reopen4', 'reopen4', 'reopen5', 'reopen6', 'exit1', 'exit2', 'exit3', 'exit4', 'exit5', 'exit6'])
```

```python
days_only_df = days_only_df[['days1', 'days2', 'days3', 'days4', 'days5' , 'days6']].fillna(pd.Timedelta(days =0))
```
