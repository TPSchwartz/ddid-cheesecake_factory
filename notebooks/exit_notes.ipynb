```python
import numpy as np
import pandas as pd
```

```python
exit_notes_df = pd.read_excel('..\data\TEIS Exit Data for NSS.xlsx', sheet_name = 'Exit Notes')
```

```python
reopened_df = exit_notes_df[exit_notes_df['Exit Note'].str.contains('reopened', case=False, na=False)].reset_index(drop=True)
reopened_df = reopened_df.drop(columns=['POE', 'DOB', 'Notification Date', 'Initial Eligibility Date'])
reopened_df
```

```python
# finding all dates
dates_df = reopened_df['Exit Note'].str.findall('(\\d{2}/\\d{2}/\\d{4})')
dates_df = dates_df.to_frame()
dates_df['total_dates'] = dates_df['Exit Note'].apply(lambda x: len(str(x).split(' ')))
dates_df
```

```python
# finding the reopened and exit dates
reopen_date_df = reopened_df['Exit Note'].str.findall('Reopened: \\d{2}/\\d{2}/\\d{4}')
exit_date_df = reopened_df['Exit Note'].str.findall('Exit Date: \\d{2}/\\d{2}/\\d{4}')
```

```python
# pull dates
reopened_df['reopened_dates'] = reopened_df['Exit Note'].str.findall('Reopened: \\d{2}/\\d{2}/\\d{4}')
reopened_df['exit_dates'] = reopened_df['Exit Note'].str.findall('Exit Date: \\d{2}/\\d{2}/\\d{4}')
# convert to string
reopen_date_df = reopened_df.reopened_dates.to_frame().astype(str)
exit_date_df = reopened_df.exit_dates.to_frame().astype(str)
# get each date in its own column
reopen_date_df = reopen_date_df.reopened_dates.str.split(',', expand=True)
reopen_date_df.columns = ['reopen1', 'reopen2', 'reopen3', 'reopen4', 'reopen5', 'reopen6']
exit_date_df = exit_date_df.exit_dates.str.split(',', expand=True)
exit_date_df.columns = ['exit1', 'exit2', 'exit3', 'exit4', 'exit5', 'exit6']

cols_r = ['reopen1', 'reopen2', 'reopen3', 'reopen4', 'reopen5', 'reopen6']
cols_e = ['exit1', 'exit2', 'exit3', 'exit4', 'exit5', 'exit6']
# trim unnecessary characters from dates
reopen_date_df[cols_r] = reopen_date_df[cols_r].replace({'\[':'', '\]':'', '\'':'', 'Reopened:':''}, regex=True)
exit_date_df[cols_e] = exit_date_df[cols_e].replace({'\[':'', '\]':'', '\'':'', 'Exit Date:':''}, regex=True)
display(reopen_date_df)
display(exit_date_df)
```

```python
reopen_exit_df = reopen_date_df.join(exit_date_df)
reopen_exit_df = reopen_exit_df[reopen_exit_df.columns[[6,0,7,1,8,2,9,3,10,4,11,5]]]
reopen_exit_df = reopen_exit_df.astype('datetime64')
reopen_exit_df
```

```python
exit_notes_stripped = reopened_df.drop(columns = ['Latest IFSP Date', 'Child ID', 'Exit Reason', 'Exit Note', 'reopened_dates', 'exit_dates'])
exit_notes_stripped
```

```python
reopen_exit_notes = exit_notes_stripped.join(reopen_exit_df)
reopen_exit_notes
```

```python
# getting rid of the dates before Initial IFSP Date, because they dont matter
reopen_exit_notes.loc[reopen_exit_notes.exit1 < reopen_exit_notes['Initial IFSP Date'], ['exit1']] = 'NaT'
reopen_exit_notes.loc[reopen_exit_notes.exit2 < reopen_exit_notes['Initial IFSP Date'], ['exit2']] = 'NaT'
reopen_exit_notes.loc[reopen_exit_notes.exit3 < reopen_exit_notes['Initial IFSP Date'], ['exit3']] = 'NaT'
reopen_exit_notes.loc[reopen_exit_notes.exit4 < reopen_exit_notes['Initial IFSP Date'], ['exit4']] = 'NaT'
reopen_exit_notes.loc[reopen_exit_notes.exit5 < reopen_exit_notes['Initial IFSP Date'], ['exit5']] = 'NaT'
reopen_exit_notes.loc[reopen_exit_notes.exit6 < reopen_exit_notes['Initial IFSP Date'], ['exit6']] = 'NaT'
reopen_exit_notes.loc[reopen_exit_notes.reopen1 < reopen_exit_notes['Initial IFSP Date'], ['reopen1']] = 'NaT'
reopen_exit_notes.loc[reopen_exit_notes.reopen2 < reopen_exit_notes['Initial IFSP Date'], ['reopen2']] = 'NaT'
reopen_exit_notes.loc[reopen_exit_notes.reopen3 < reopen_exit_notes['Initial IFSP Date'], ['reopen3']] = 'NaT'
reopen_exit_notes.loc[reopen_exit_notes.reopen4 < reopen_exit_notes['Initial IFSP Date'], ['reopen4']] = 'NaT'
reopen_exit_notes.loc[reopen_exit_notes.reopen5 < reopen_exit_notes['Initial IFSP Date'], ['reopen5']] = 'NaT'
reopen_exit_notes.loc[reopen_exit_notes.reopen6 < reopen_exit_notes['Initial IFSP Date'], ['reopen6']] = 'NaT'

reopen_exit_notes
```

```python
reopen_exit_notes = reopen_exit_notes[reopen_exit_notes.exit1.notnull()].reset_index(drop=True)

reopen_exit_notes.exit1 = np.where(reopen_exit_notes.exit1 == reopen_exit_notes['Exit Date'], reopen_exit_notes.exit2, reopen_exit_notes.exit1)
```

```python
# subtracting the days
reopen_exit_notes['r1-e1'] = reopen_exit_notes.reopen1 - reopen_exit_notes.exit1
reopen_exit_notes['r2-e2'] = reopen_exit_notes.reopen2 - reopen_exit_notes.exit2
reopen_exit_notes['r3-e3'] = reopen_exit_notes.reopen3 - reopen_exit_notes.exit3
reopen_exit_notes['r4-e4'] = reopen_exit_notes.reopen4 - reopen_exit_notes.exit4
reopen_exit_notes['r5-e5'] = reopen_exit_notes.reopen5 - reopen_exit_notes.exit5
reopen_exit_notes['r6-e6'] = reopen_exit_notes.reopen6 - reopen_exit_notes.exit6
```

```python
reopen_exit_notes['r1-e1'] = reopen_exit_notes['r1-e1'].fillna(pd.Timedelta(seconds=0))
reopen_exit_notes['r2-e2'] = reopen_exit_notes['r2-e2'].fillna(pd.Timedelta(seconds=0))
reopen_exit_notes['r3-e3'] = reopen_exit_notes['r3-e3'].fillna(pd.Timedelta(seconds=0))
reopen_exit_notes['r4-e4'] = reopen_exit_notes['r4-e4'].fillna(pd.Timedelta(seconds=0))
reopen_exit_notes['r5-e5'] = reopen_exit_notes['r5-e5'].fillna(pd.Timedelta(seconds=0))
reopen_exit_notes['r6-e6'] = reopen_exit_notes['r6-e6'].fillna(pd.Timedelta(seconds=0))
```

```python
reopen_exit_notes['day_diff'] = reopen_exit_notes['r1-e1'] + reopen_exit_notes['r2-e2'] + reopen_exit_notes['r3-e3'] + reopen_exit_notes['r4-e4'] + reopen_exit_notes['r5-e5'] + reopen_exit_notes['r6-e6']
reopen_exit_notes.day_diff = reopen_exit_notes.day_diff.astype('timedelta64[D]').astype(np.int64)
reopen_exit_notes['actual_service_time'] = reopen_exit_notes['Days from I-IFSP to Exit'].astype('timedelta64[D]').astype(np.int64) - reopen_exit_notes.day_diff
```

```python
service_time_df = reopen_exit_notes[['actual_service_time']]
service_time_df
```
